```python
import tkinter as tk
from tkinter import ttk, messagebox
from PIL import Image, ImageTk
import joblib
import os

# Load the XGBoost model
model_path = "C:/Users/asus1/Desktop/XGB.joblib"
try:
    model = joblib.load(model_path)
except FileNotFoundError:
    messagebox.showerror("Model Error", "Model file not found. Please ensure the path is correct.")
    exit()

def predict():
    try:
        x1 = float(entry_x1.get())
        x2 = float(entry_x2.get())
        x3 = float(entry_x3.get())
        x4 = float(entry_x4.get())
        x5 = float(entry_x5.get())

        prediction = model.predict([[x1, x2, x3, x4, x5]])[0]
        output_label.config(text=f"{prediction:.4f}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numbers for all inputs.")
    except Exception as e:
        messagebox.showerror("Prediction Error", f"An error occurred: {e}")

# Create main window
root = tk.Tk()
root.title("SWI Wedge Length Ratio Prediction Model")
root.geometry("700x850")

title_label = tk.Label(root, text="Modeling the Impact of Groundwater Abstraction and Concrete Dam Fractures on\n Saltwater Intrusion Using Interpretable Machine Learning Models", font=("Arial", 12, "bold"))
title_label.pack(fill="x", padx=10, pady=10)

developers_label = tk.Label(root, text="Developers: Mohamed Kamel Elshaarawy & Asaad Mater Armanuos", font=("Arial", 12), wraplength=700, justify="center")
developers_label.pack(fill="x", padx=10, pady=5)

# Add image
image_path = "C:/Users/asus1/Desktop/sketch.png"
if os.path.exists(image_path):
    try:
        image = Image.open(image_path)
        image = image.resize((480, 250), Image.LANCZOS)
        photo = ImageTk.PhotoImage(image)
        image_label = tk.Label(root, image=photo)
        image_label.image = photo
        image_label.pack(padx=5, pady=5)
    except Exception as e:
        messagebox.showwarning("Image Error", f"Unable to load image: {e}")
else:
    messagebox.showwarning("Image Missing", "The sketch image file is missing. The application will run without it.")

# Input frame
input_frame = ttk.LabelFrame(root, text="Input Parameters (Dimensionless Terms)", padding=(5, 5, 5, 5))
input_frame.pack(padx=10, pady=10, fill="x")

# Define input fields with LaTeX-style variable labels
ttk.Label(input_frame, text=r"Relative fracture aperture height (Hf/La):", font=("Arial", 12, "bold")).grid(row=0, column=0, sticky=tk.W, padx=5, pady=5)
entry_x1 = ttk.Entry(input_frame, font=("Arial", 14))
entry_x1.grid(row=0, column=1, padx=5, pady=5)

ttk.Label(input_frame, text=r"Relative fracture aperture diameter (Df/La):", font=("Arial", 12, "bold")).grid(row=1, column=0, sticky=tk.W, padx=5, pady=5)
entry_x2 = ttk.Entry(input_frame, font=("Arial", 14))
entry_x2.grid(row=1, column=1, padx=5, pady=5)

ttk.Label(input_frame, text=r"Relative well height (Hw/La):", font=("Arial", 12, "bold")).grid(row=2, column=0, sticky=tk.W, padx=5, pady=5)
entry_x3 = ttk.Entry(input_frame, font=("Arial", 14))
entry_x3.grid(row=2, column=1, padx=5, pady=5)

ttk.Label(input_frame, text=r"Relative well distance (Lw/La):", font=("Arial", 12, "bold")).grid(row=3, column=0, sticky=tk.W, padx=5, pady=5)
entry_x4 = ttk.Entry(input_frame, font=("Arial", 14))
entry_x4.grid(row=3, column=1, padx=5, pady=5)

ttk.Label(input_frame, text=r"Relative well abstraction rate (Q/(gLa)^0.5):", font=("Arial", 12, "bold")).grid(row=4, column=0, sticky=tk.W, padx=5, pady=5)
entry_x5 = ttk.Entry(input_frame, font=("Arial", 14))
entry_x5.grid(row=4, column=1, padx=5, pady=5)

# Output
output_frame = ttk.LabelFrame(root, text="Prediction Result", padding=(5, 5, 5, 5))
output_frame.pack(padx=10, pady=10, fill="x")

ttk.Label(output_frame, text=r"Predicted Relative SWI wedge length (L/La):", font=("Arial", 12, "bold")).grid(row=0, column=0, padx=5, pady=5, sticky=tk.W)
output_label = ttk.Label(output_frame, text="", font=("Arial", 16), foreground="red")
output_label.grid(row=0, column=1, padx=5, pady=5)

# Buttons
buttons_frame = tk.Frame(root)
buttons_frame.pack(pady=10)

predict_button = ttk.Button(buttons_frame, text="Predict", command=predict)
predict_button.grid(row=0, column=0, padx=5)

clear_button = ttk.Button(buttons_frame, text="Clear", command=lambda: [
    entry_x1.delete(0, tk.END),
    entry_x2.delete(0, tk.END),
    entry_x3.delete(0, tk.END),
    entry_x4.delete(0, tk.END),
    entry_x5.delete(0, tk.END),
    output_label.config(text="")
])
clear_button.grid(row=0, column=1, padx=5)

# Run the application
root.mainloop()
```
